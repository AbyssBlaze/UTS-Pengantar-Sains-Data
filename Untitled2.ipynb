```python
#Import the libraries
from sklearn.svm import SVR
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn-darkgrid')
```

```python
#Load the data
from google.colab import files
uploaded = files.upload()
```

```python
#Store and look at the data
df = pd.read_csv('GOOG_stok.csv')
df
```

```python
#Get and print the last row
actual_price = df.tail(1)
actual_price
```

```python
#Prepare the data for training
df = df.head(len(df)-1)
df
```

```python
#Create and empty list to store the independent and the dependent data
days = list()
adj_close_prices = list()
```

```python
#Get the date and the adjusted close price
df_days = df.loc[:, 'Date']
df_adj_close = df.loc[:, 'Adj Close']
```

```python
#Create the independent data set
for day in df_days:
   days.append( [int(day.split('/')[1])] )

#Create the dependent data set
for adj_close in df_adj_close:
   adj_close.append( float(adj_close) )
```

```python
print(days)
```

```python
#Create and train an SVR model using a linear kernel
lin_svr = SVR(kernel='linear', C=1000.0)
lin_svr.fit(days, adj_close)

#Create and train an SVR model using a polynomial kernel
poly_svr = SVR(kernel='poly', C=1000.0, degree=2)
poly_svr.fit(days, adj_close)

#Create and train an SVR model using a RBF kernel
rbf_svr = SVR(kernel='rbf', C=1000.0, gamma=0.15)
rbf_svr.fit(days, adj_close)
```

```python
#Plot the models on a graph to see which has the best fit
plt.figure(figsize=(16,8))
plt.scatter(days, adj_close_prices, color = 'black', label='Original Data')
plt.plot(days, rbf_svr.predict(days), color = 'green', label='RBF Model')
plt.plot(days, poly_svr.predict(days), color = 'orange', label='Polynomial Model')
plt.plot(days, lin_svr.predict(days), color = 'purple', label='Linear Model')
plt.xlabel('Days')
plt.ylabel('Adj Close Price')
plt.title('Support Vector Regression')
plt.legend()
plt.show()
```
